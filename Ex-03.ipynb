```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)
x = np.linspace(0, 10, 100).reshape(-1, 1)
y = 3 * x**2 + 2 * x + np.random.randn(100, 1) * 10
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
y_pred_linear = linear_model.predict(x_test)
```

```python
poly_features = PolynomialFeatures(degree=2)
x_poly_train = poly_features.fit_transform(x_train)
x_poly_test = poly_features.transform(x_test)
```

```python
poly_model = LinearRegression()
poly_model.fit(x_poly_train, y_train)
y_pred_poly = poly_model.predict(x_poly_test)
```

```python
plt.scatter(x_test, y_test, color='blue', label='Actual Data')
plt.scatter(x_test, y_pred_linear, color='red', label='Linear Regression', alpha=0.6)
plt.scatter(x_test, y_pred_poly, color='green', label='Polynomial Regression', alpha=0.6)
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear vs Polynomial Regression")
plt.show()
```

```python
print(f"Linear Regression R^2 Score: {linear_model.score(x_test, y_test):.4f}")
print(f"Polynomial Regression R^2 Score: {poly_model.score(x_poly_test, y_test):.4f}")
```
